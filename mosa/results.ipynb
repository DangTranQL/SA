```python
import numpy as np
import matplotlib.pyplot as plt
```

# Negative Autoregulation

```python
run32_1 = np.load('batch32_iter1/mosa_sobol_neg_200.npz')
run32_10 = np.load('batch32_iter10/mosa_sobol_neg_200.npz')
run1024_1 = np.load('batch1024_iter1/mosa_sobol_neg_200.npz')
run1024_10 = np.load('batch1024_iter10/mosa_sobol_neg_200.npz')
```

```python
data = np.load('data/ARneg_SensitivityPareto.npy')

metric32_1 = np.load('batch32_iter1/mosa_sobol_neg_metrics.npz')
metric32_10 = np.load('batch32_iter10/mosa_sobol_neg_metrics.npz')
metric1024_1 = np.load('batch1024_iter1/mosa_sobol_neg_metrics.npz')
metric1024_10 = np.load('batch1024_iter10/mosa_sobol_neg_metrics.npz')
```

```python
fig, ax1 = plt.subplots(figsize=(8, 8))

ax1.scatter(data[:, 0], data[:, 1], label='Grid Search', color='black', s=2, zorder=5)
ax1.scatter(run32_1['f1'], run32_1['f2'], label='Run 32 Iter 1', color='green', s=5, zorder=4)
ax1.scatter(run32_10['f1'], run32_10['f2'], label='Run 32 Iter 10', color='#87CEFA', s=20, zorder=3)
ax1.scatter(run1024_1['f1'], run1024_1['f2'], label='Run 1024 Iter 1', color='#9932CC', s=39, zorder=2)
ax1.scatter(run1024_10['f1'], run1024_10['f2'], label='Run 1024 Iter 10', color='red', s=70, zorder=1)
ax1.legend()

ax1.set_xlabel(r'$|S_{\alpha} (x_{ss})|$')
ax1.set_ylabel(r'$|S_{n} (x_{ss})|$')
ax1.set_title("Negative Autoregulation: Pareto Front Evolution")

plt.tight_layout()
plt.show()
```

```python
plt.rcParams['font.weight'] = 'bold'
plt.rcParams['axes.titleweight'] = 'bold'
plt.rcParams['axes.titlesize'] = 20
plt.rcParams['axes.labelweight'] = 'bold'
plt.rcParams['axes.titleweight'] = 'bold'
plt.rcParams['xtick.labelsize'] = 14
plt.rcParams['ytick.labelsize'] = 14
plt.rcParams['axes.linewidth'] = 2

fig, axs = plt.subplots(4, 3, figsize=(36, 48))

axs[0, 0].plot(metric32_1['gd'], linewidth=3)
axs[0, 0].set_title("GD (32_1)")

axs[0, 1].plot(metric32_1['igd'], linewidth=3)
axs[0, 1].set_title("IGD (32_1)")

axs[0, 2].plot(metric32_1['hv'], linewidth=3)
axs[0, 2].set_title("HV (32_1)")

axs[1, 0].plot(metric32_10['gd'], linewidth=3)
axs[1, 0].set_title("GD (32_10)")

axs[1, 1].plot(metric32_10['igd'], linewidth=3)
axs[1, 1].set_title("IGD (32_10)")

axs[1, 2].plot(metric32_10['hv'], linewidth=3)
axs[1, 2].set_title("HV (32_10)")

axs[2, 0].plot(metric1024_1['gd'], linewidth=3)
axs[2, 0].set_title("GD (1024_1)")

axs[2, 1].plot(metric1024_1['igd'], linewidth=3)
axs[2, 1].set_title("IGD (1024_1)")

axs[2, 2].plot(metric1024_1['hv'], linewidth=3)
axs[2, 2].set_title("HV (1024_1)")

axs[3, 0].plot(metric1024_10['gd'], linewidth=3)
axs[3, 0].set_title("GD (1024_10)")

axs[3, 1].plot(metric1024_10['igd'], linewidth=3)
axs[3, 1].set_title("IGD (1024_10)")

axs[3, 2].plot(metric1024_10['hv'], linewidth=3)
axs[3, 2].set_title("HV (1024_10)")

plt.tight_layout(h_pad=2.0)
plt.show()
```

# Positive-Negative Feedback

```python
data = np.load('data/PosNegFbGridSearchData_renamed.npz')

metric32_1 = np.load('batch32_iter1/mosa_sobol_posneg_metrics.npz')
metric32_10 = np.load('batch32_iter10/mosa_sobol_posneg_metrics.npz')
metric1024_1 = np.load('batch1024_iter1/mosa_sobol_posneg_metrics.npz')
metric1024_10 = np.load('batch1024_iter10/mosa_sobol_posneg_metrics.npz')
```
