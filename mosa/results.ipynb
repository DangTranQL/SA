```python
# NEG
time32_1_neg = 2
time32_10_neg = 22
time1024_1_neg = 66
time1024_10_neg = 642
time_4096_1_neg = 258
time_4096_1_adapt_neg = 252

run32_1_neg = (metric32_1_neg['gd'][-1] + metric32_1_neg['igd'][-1] + 1-metric32_1_neg['hv'][-1]) / 3
run32_10_neg = (metric32_10_neg['gd'][-1] + metric32_10_neg['igd'][-1] + 1-metric32_10_neg['hv'][-1]) / 3
run1024_1_neg = (metric1024_1_neg['gd'][-1] + metric1024_1_neg['igd'][-1] + 1-metric1024_1_neg['hv'][-1]) / 3
run1024_10_neg = (metric1024_10_neg['gd'][-1] + metric1024_10_neg['igd'][-1] + 1-metric1024_10_neg['hv'][-1]) / 3
run4096_1_neg = (metric4096_1_neg['gd'][-1] + metric4096_1_neg['igd'][-1] + 1-metric4096_1_neg['hv'][-1]) / 3
run4096_1_adapt_neg = (metric4096_1_adapt_neg['gd'][-1] + metric4096_1_adapt_neg['igd'][-1] + 1-metric4096_1_adapt_neg['hv'][-1]) / 3

#POSNEG
time32_1_posneg = 6
time32_10_posneg = 50
time1024_1_posneg = 172
time1024_10_posneg = 1404
time_4096_1_posneg = 639
time_4096_1_adapt_posneg = 627

run32_1_posneg = (metric32_1_posneg['gd'][-1] + metric32_1_posneg['igd'][-1] + 1-metric32_1_posneg['hv'][-1]) / 3
run32_10_posneg = (metric32_10_posneg['gd'][-1] + metric32_10_posneg['igd'][-1] + 1-metric32_10_posneg['hv'][-1]) / 3
run1024_1_posneg = (metric1024_1_posneg['gd'][-1] + metric1024_1_posneg['igd'][-1] + 1-metric1024_1_posneg['hv'][-1]) / 3
run1024_10_posneg = (metric1024_10_posneg['gd'][-1] + metric1024_10_posneg['igd'][-1] + 1-metric1024_10_posneg['hv'][-1]) / 3
run4096_1_posneg = (metric4096_1_posneg['gd'][-1] + metric4096_1_posneg['igd'][-1] + 1-metric4096_1_posneg['hv'][-1]) / 3
run4096_1_adapt_posneg = (metric4096_1_adapt_posneg['gd'][-1] + metric4096_1_adapt_posneg['igd'][-1] + 1-metric4096_1_adapt_posneg['hv'][-1]) / 3
```

```python
x = [run32_1_neg, run32_10_neg, run1024_1_neg, run1024_10_neg, run4096_1_neg, run4096_1_adapt_neg,
     run32_1_posneg, run32_10_posneg, run1024_1_posneg, run1024_10_posneg, run4096_1_posneg, run4096_1_adapt_posneg]

y = [time32_1_neg, time32_10_neg, time1024_1_neg, time1024_10_neg, time_4096_1_neg, time_4096_1_adapt_neg,
     time32_1_posneg, time32_10_posneg, time1024_1_posneg, time1024_10_posneg, time_4096_1_posneg, time_4096_1_adapt_posneg]

labels = ['32_1', '32_10', '1024_1', '1024_10', '4096_1', '4096_1_adapt']

markers = ['o', 's', '^', 'D', 'v', 'P']

plt.figure(figsize=(10,6))

for i, label in enumerate(labels):
    # Negative autoregulation (same color for all neg)
    plt.scatter(x[i], y[i], color='tab:blue', marker=markers[i], s=100, label=f'{label} Neg')
    # Positive-negative feedback (same color for all posneg)
    plt.scatter(x[i+6], y[i+6], color='tab:orange', marker=markers[i], s=100, label=f'{label} PosNeg')

plt.xlabel('Avg(GD, IGD, HV)')
plt.ylabel('Runtime')
plt.title('Runtime vs Avg Metrics')

# Optional: simplify legend (one per run with marker, color indicates type)
from matplotlib.lines import Line2D

legend_elements = []
for i, label in enumerate(labels):
    legend_elements.append(Line2D([0], [0], marker=markers[i], color='tab:blue', label=f'{label} Neg',
                                  markersize=8, linestyle=''))
    legend_elements.append(Line2D([0], [0], marker=markers[i], color='tab:orange', label=f'{label} PosNeg',
                                  markersize=8, linestyle=''))

plt.legend(handles=legend_elements, bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.tight_layout()
plt.show()
```
